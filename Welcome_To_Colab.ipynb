```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```
df=pd.read_csv(r"/content/Iris (1).csv")
df.head()
```

```
df.shape
```

```
df.isnull().sum()
```

```
df.columns
```

```


plt.figure(figsize=(6, 4))

plt.hist(df['SepalLengthCm'], bins=20, alpha=0.6, label='SepalLengthCm')
plt.hist(df['PetalLengthCm'], bins=20, alpha=0.6, label='PetalLengthCm')

plt.xlabel("Length (cm)")
plt.ylabel("Frequency")
plt.title("Histogram of SepalLength and PetalLength")
plt.legend()

plt.show()
```

```
import matplotlib.pyplot as plt

# Select numeric columns
cols = df.select_dtypes(include="number").columns.drop('Id')

# Plot one histogram at a time
for col in cols:
    plt.figure(figsize=(6, 4))
    plt.hist(df[col], bins=20, color='skyblue')
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
```

```
import matplotlib.pyplot as plt

# Count of each species
species_counts = df["Species"].value_counts()

plt.figure(figsize=(6,4))
plt.bar(species_counts.index, species_counts.values, color="blue")
plt.title("Species Distribution")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```
